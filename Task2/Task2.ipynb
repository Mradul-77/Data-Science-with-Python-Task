## Importing and connecting the .csv file

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('01.Data Cleaning and Preprocessing.csv')

# printing few rows to check whether the csv is loaded or not

df.head()
```

## Handling missing values

```python
# checking for cells with missing values

df.isnull()
```

```python
# total number of missing values in each column

df.isnull().sum()
```

```python
# filling the missing values with 0

df_cleaned=df.fillna(0)
```

```python
# checking for null values in each column again after cleaning

df_cleaned.isnull().sum()
```

```python
# cleaned data frame

df_cleaned
```

## Filtering data based on any condition

```python
# 1. let's take a condition in which we want all the rows and columns where ChipRate > 15.000

df_filtered=df_cleaned[df_cleaned['ChipRate'] > 15.000]
df_filtered
```

```python
# 2. Let's now select and filter all the columns who's name is starting with the letter 'C'

df_filtered_columns=df_cleaned.filter(regex='^C')
df_filtered_columns
```

## Calculating Summary Statistics

```python
# generating descriptive statistics

df_cleaned.describe()
```

```python
# let's calculate median values of each column

df_median=df_cleaned.drop('Observation', axis='columns').median(axis='index')
df_median
```

```python
# let's calculate mode values of each column

df_mode=df_cleaned.drop('Observation', axis='columns').mode(axis='index').iloc[0]
df_mode
```

```python
# Calculating the correlation matrix

df_correlation = df_cleaned.drop('Observation', axis='columns').corr()
df_correlation
```
